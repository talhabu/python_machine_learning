# One Hot Encoder using Loop

```python
# Using Loop
for col in df4.columns:
    if is_numeric_dtype(df4[col]):
        continue
    else:
        one = pd.get_dummies(df4[col])
        df4.drop(df4[[col]], axis=1, inplace=True)
        df4 = pd.concat([df4, one], axis=1)
```

```python
df4.head()
```

# Ordinal Encoder Using For Loop

```python
df9.head()
```

```python
cols = ['sex', 'smoker', 'region']
for col in cols:
    unique = df9[col].unique()
    df9[col] = OrdinalEncoder(categories=[unique]).fit_transform(df9[[col]])
```
