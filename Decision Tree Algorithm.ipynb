# Preparing The Data

```python
#import neccessary Library
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/abutalha/Documents/Data Analyst/AI Quest/Data Science/game data.csv')
```

```python
df.head()
```

```python
df.drop('Days', axis=1, inplace=True )
```

```python
df.head()
```

```python
x = df.drop('Wear Jacket?', axis=1)
```

```python
x.head()
```

```python
y = df['Wear Jacket?']
```

# Encoding the data

```python
#import Encding Library
from sklearn.preprocessing import LabelEncoder
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
le = LabelEncoder()
```

```python
#Using for loop to encoder all the features together 

for col in x.columns:
    if is_numeric_dtype(x[col]):
        continue
    x[col] = le.fit_transform(x[col])
```

```python
x.head()
```

# Decsition Tree Classifier Algorithm

```python
#Import Library
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
#fit the model 
clf.fit(x,y)
```

```python
#Predict the result
clf.predict([[1,0,1]])
```

```python
clf.predict([[0,0,0]])
```
